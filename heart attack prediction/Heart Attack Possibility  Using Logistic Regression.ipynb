## Assignment Of Heart Attack Possibility  Using Logistic Regression

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Reading the csv file

```python
flatdata = pd.read_csv('heart.csv')
flatdata
```

```python
X = flatdata.iloc [:,0:13]# independent
y = flatdata.iloc [:,13]# dependent


X
```

```python
X.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()# create object of LE
X["age"] = lb.fit_transform(X["age"])
#print (X)

X
```

```python
from sklearn.impute import SimpleImputer
# First create an Imputer , Stratergy means what we want to write in place of missed value
mss = SimpleImputer (missing_values = np.nan, strategy = 'mean')  
 
# Set which columns imputer should perform
mss = mss.fit (X.iloc[:,1:13])
# update values of X with new values
X.iloc[:,1:13] = mss.transform(X.iloc[:,1:13])

print(X)
```

```python


from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = .10, random_state = 0)


#print("splitting done")
#print(X_train)		
#print(y_train)	
#print(X_test)	
X_train
X_test
```

```python
from sklearn.linear_model import LogisticRegression
LRClassifier = LogisticRegression ()# object  of algorithm
LRClassifier.fit (X_train, y_train)

prediction = LRClassifier.predict (X_test)
print(prediction)
print(y_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, prediction))
```
